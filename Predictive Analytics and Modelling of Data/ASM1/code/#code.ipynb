# Read file


```python
import pandas as pd
from io import StringIO

with open("../data/CS_data.txt", "r", encoding="utf-8") as file:
    data = file.read()

# Load the CS_data.txt file into a DataFrame
data_df = pd.read_csv(StringIO(data), sep="\t")
print(type(data_df))

# Display the first few rows of the dataset to understand its structure
data_df.head()
print(data_df.columns)
```

# data cleaning and pre-processing


```python
# Display descriptive statistics of the dataset
data_descriptive_stats = data_df.describe(include="all")

# Display the descriptive statistics for the first few columns for brevity
data_descriptive_stats.iloc[:, :10]
```

### handle missing values
identifying all the missing values according to variable list

```python
# Re-reading the variable list from the provided Excel file
variable_list_df = pd.read_excel("../data/variable list.XLS.xls")

# Display the content of the variable list
variable_list_df[["Var_Title", "Var_Description", "Field_Content"]]
```

```python
# variable_list_df.to_json("variable_list_df.json")
# data_df.to_json("data_df.json")
```

```python
variable_list_df["Var_Title"]
for i in data_df.columns:
    if i in [
        # "OTHER_INCOMES",
        # "QUANT_BANKING_ACCOUNTS",
        # "QUANT_SPECIAL_BANKING_ACCOUNTS",
        # "PERSONAL_ASSETS_VALUE",
        # "QUANT_CARS",
        # "COMPANY",
        "PAYMENT_DAY",
    ]:
        print("\n", data_df[i].value_counts().to_string())
        indices_for_value = variable_list_df["Var_Title"][
            variable_list_df["Var_Title"] == i
        ].index
        print(
            "\n",
            variable_list_df["Field_Content"][indices_for_value].to_string(),
            "\n----------------------",
        )
```

```python
# Calculate the number of missing values for each column
missing_values = data_df.isnull().sum()

# Display columns with missing values and their counts
print(missing_values[missing_values > 0])

# Calculate the percentage of missing values for each column
missing_percentage = (data_df.isnull().sum() / len(data_df)) * 100
print(
    "----------------------------------------------\n",
    missing_percentage[missing_percentage > 50],
)
```
