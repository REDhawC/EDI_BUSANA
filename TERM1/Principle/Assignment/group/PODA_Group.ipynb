```python
import matplotlib.pyplot as plt

num_but_cat = [col for col in df.columns if df[col].nunique() < 6 and df[col].dtypes != "O"]

num_cols = [col for col in df.columns if df[col].dtypes != "O"]
num_cols = [col for col in num_cols if col not in num_but_cat]

df[num_cols].hist(bins=30, figsize=(15,15))
plt.show()
```

```python
satisfied_employees1 = df[df['JobSatisfaction'] == 1]['MonthlyIncome']
satisfied_employees2 = df[df['JobSatisfaction'] == 2]['MonthlyIncome']
satisfied_employees3 = df[df['JobSatisfaction'] == 3]['MonthlyIncome']
satisfied_employees4 = df[df['JobSatisfaction'] == 4]['MonthlyIncome']


# print(len(satisfied_employees1))
# print(len(satisfied_employees2))
# print(len(satisfied_employees3))
# print(len(satisfied_employees4))
```

```python
from scipy import stats

p_value1 = stats.shapiro(satisfied_employees1)[1]
p_value2 = stats.shapiro(satisfied_employees2)[1]
p_value3 = stats.shapiro(satisfied_employees3)[1]
p_value4 = stats.shapiro(satisfied_employees4)[1]

if all(p > 0.05 for p in [p_value1, p_value2, p_value3, p_value4]):
    print("所有数据组看起来符合正态分布。")
else:
    print("至少有一组数据不符合正态分布。")
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

data = [satisfied_employees1, satisfied_employees2, satisfied_employees3, satisfied_employees4]

for i, dataset in enumerate(data, 1):
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.hist(dataset, bins=20)
    plt.title(f'Histogram of satisfied_employees {i}')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(dataset, line='s', ax=plt.gca())
    plt.title(f'Q-Q Plot of satisfied_employees {i}')
    
    plt.show()
```

```python
from scipy.stats import levene

p_value = levene(satisfied_employees1, satisfied_employees2, satisfied_employees3, satisfied_employees4)[1]

if p_value > 0.05:
    print("数据具有相等的方差（方差齐性）。")
else:
    print("数据具有不相等的方差（方差不齐性）。")
```

```python
from scipy import stats

statistic, p_value = stats.kruskal(*data)

# 输出检验结果
print("Kruskal-Wallis统计量:", statistic)
print("p-值:", p_value)

if p_value < 0.05:
    print("存在显著差异，至少有一组的中位数不同。")
else:
    print("没有显著差异，各组中位数相似。")
```

```python
BusinessTravel1 = df[df['BusinessTravel'] == 'Non-Travel']['MonthlyIncome']
BusinessTravel2 = df[df['BusinessTravel'] == 'Travel_Rarely']['MonthlyIncome']
BusinessTravel3 = df[df['BusinessTravel'] == 'Travel_Frequently']['MonthlyIncome']
```

```python
data1 = [BusinessTravel1, BusinessTravel2, BusinessTravel3]

for i, dataset in enumerate(data1, 1):
    plt.figure(figsize=(12, 3))
    
    plt.subplot(1, 2, 1)
    plt.hist(dataset, bins=20)
    plt.title(f'Histogram of Business Travel {i}')
    
    plt.subplot(1, 2, 2)
    sm.qqplot(dataset, line='s', ax=plt.gca())
    plt.title(f'Q-Q Plot of Business Travel {i}')
    
    plt.show()
```

```python
p_value1 = levene(BusinessTravel1, BusinessTravel2, BusinessTravel3)[1]

if p_value1 > 0.05:
    print("数据具有相等的方差（方差齐性）。")
else:
    print("数据具有不相等的方差（方差不齐性）。")
```

```python
statistic1, p_value1 = stats.kruskal(*data1)

# 输出检验结果
print("Kruskal-Wallis统计量:", statistic1)
print("p-值:", p_value1)

if p_value1 < 0.05:
    print("存在显著差异，至少有一组的中位数不同。")
else:
    print("没有显著差异，各组中位数相似。")
```

```python
TotalWorkingYears1 = df[df['TotalWorkingYears'] <= 9]['MonthlyIncome']
TotalWorkingYears2 = df[(df['TotalWorkingYears'] >= 10) & (df['TotalWorkingYears'] <= 19)]['MonthlyIncome']
TotalWorkingYears3 = df[(df['TotalWorkingYears'] >= 20) & (df['TotalWorkingYears'] <= 29)]['MonthlyIncome']
TotalWorkingYears4 = df[(df['TotalWorkingYears'] >= 30) & (df['TotalWorkingYears'] <= 40)]['MonthlyIncome']

p_value2 = levene(TotalWorkingYears1, TotalWorkingYears2, TotalWorkingYears3, TotalWorkingYears4)[1]
data2 = [TotalWorkingYears1, TotalWorkingYears2, TotalWorkingYears3, TotalWorkingYears4]
statistic2, p_value2 = stats.kruskal(*data2)

# 输出检验结果
print("Kruskal-Wallis统计量:", statistic2)
print("p-值:", p_value2)

if p_value2 < 0.05:
    print("存在显著差异，至少有一组的中位数不同。")
else:
    print("没有显著差异，各组中位数相似。")
```

```python
from scipy import stats
from statsmodels.sandbox.stats.multicomp import multipletests

dunn_results = []

# 对四组数据进行两两比较
for i in range(len(data2)):
    for j in range(i+1, len(data2)):
        group1 = data2[i]
        group2 = data2[j]
        dunn_result = stats.kruskal(group1, group2)
        dunn_results.append((i+1, j+1, dunn_result.pvalue))

# 进行多重比较校正
_, dunn_results_corrected, _, _ = multipletests([p for i, j, p in dunn_results], method='bonferroni')

# 打印Dunn's检验结果
for (group1, group2, p_value), corrected_p in zip(dunn_results, dunn_results_corrected):
    print(f"Comparison between Group {group1} and Group {group2}: p-value = {p_value}, Corrected p-value = {corrected_p}")
```
