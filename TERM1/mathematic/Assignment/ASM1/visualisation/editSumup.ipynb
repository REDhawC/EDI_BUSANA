```python
import pandas as pd
```

```python
df = pd.read_csv("./sumup.csv")

# Round the numbers to 0 decimal places and convert to string with 'k'
for column in ["KharifProd", "RabiProd", "SummerProd"]:
    df[column] = df[column].round(0).astype(int).astype(str) + "k"

# Display the first few rows of the DataFrame
df.head()
```

```python
# # Define a function to get the top 3 crops and their production for each season
# def get_top_crops(df, season):
#     # Sort by production and select the top 3 crops
#     top_crops = df.sort_values(by=f"{season}Prod", ascending=False).head(3)
#     # Create a string with the crop names and their production
#     top_crops_str = "<br>".join(
#         [f'{row.crop}: {row[f"{season}Prod"]}' for _, row in top_crops.iterrows()]
#     )
#     return top_crops_str


# # Create a new DataFrame to store the results
# top_crops_df = pd.DataFrame(df["state"].unique(), columns=["state"])

# # Apply the function to each state and each season
# for season in ["Kharif", "Rabi", "Summer"]:
#     top_crops_df[f"{season}_top_crops"] = (
#         df.groupby("state").apply(get_top_crops, season).values
#     )

# # Display the first few rows of the new DataFrame
# top_crops_df.head()


# Define a function to get the top 3 crops and their production for each season
def get_top_crops(df, season):
    # Filter out crops with '0k' production
    df = df[df[f"{season}Prod"] != "0k"]
    # Sort by production and select the top 3 crops
    top_crops = df.sort_values(by=f"{season}Prod", ascending=False).head(3)
    # Create a string with the crop names and their production
    top_crops_str = "<br>".join(
        [f'{row.crop}: {row[f"{season}Prod"]}' for _, row in top_crops.iterrows()]
    )
    return top_crops_str


# Create a new DataFrame to store the results
top_crops_df = pd.DataFrame(df["state"].unique(), columns=["state"])

# Apply the function to each state and each season
for season in ["Kharif", "Rabi", "Summer"]:
    top_crops_df[f"{season}_top_crops"] = (
        df.groupby("state").apply(get_top_crops, season).values
    )

# Display the first few rows of the new DataFrame
top_crops_df.head()
```

```python
print(top_crops_df.shape)
```

```python
areaDf = pd.read_csv("./areaM1.csv")

areaDf.head()
```

```python
# Merge the dataframes using a left join
merged_df = pd.merge(areaDf, top_crops_df, how="left", on="state")

# Display the first few rows of the merged DataFrame
merged_df.head()
```
