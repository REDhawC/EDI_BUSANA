```python
def tranfrom_covid_days_to_quarters(original_data):
    # Define a function to convert month to quarter
    def convert_month_to_quarter(month):
        # Determine the quarter based on the month value
        if month <= 3:
            return "1"
        elif month <= 6:
            return "2"
        elif month <= 9:
            return "3"
        else:
            return "4"

    # Create a filtered list `covid_quarters` that includes data from 2020-04-01 to 2021-12-31
    covid_quarters = [
        i for i in original_data if i["date"] > 20200331 and i["date"] < 20220101
    ]

    # Add year, quarter, and time (concatenation of year and quarter) to each entry in `covid_quarters`
    for i in covid_quarters:
        i["Year"] = str(i["date"] // 10000)  # Extracting the year part from the date
        i["Quarter"] = convert_month_to_quarter(
            int(str(i["date"])[4:6])
        )  # Extracting and converting month to quarter
        i["Time"] = i["Year"] + "-" + i["Quarter"]  # Concatenating year and quarter

    # Initialize the dictionary `aggregated_data` to store aggregated data
    aggregated_data = {}

    # Iterate through the data to aggregate the required fields
    for entry in covid_quarters:
        time_key = entry["Time"]

        # If `time_key` is not present in `aggregated_data`, initialize the key with structure to hold data
        if time_key not in aggregated_data:
            aggregated_data[time_key] = {
                "whole_scotland": {
                    "new_positive": 0,
                    "new_deaths": 0,
                    "first_infections": 0,
                    "reinfections": 0,
                },
                "locations": {},
            }

        # Aggregate `whole_scotland` data
        for key in ["new_positive", "new_deaths", "first_infections", "reinfections"]:
            aggregated_data[time_key]["whole_scotland"][key] += entry["whole_scotland"][
                key
            ]

        # Aggregate data for each `location`
        for location, location_data in entry["locations"].items():
            # If `location` is not present in `locations` of `aggregated_data[time_key]`, initialize it
            if location not in aggregated_data[time_key]["locations"]:
                aggregated_data[time_key]["locations"][location] = {
                    "new_positive": 0,
                    "new_deaths": 0,
                    "first_infections": 0,
                    "reinfections": 0,
                }
            # Aggregate data for each key for the specific `location`
            for key in [
                "new_positive",
                "new_deaths",
                "first_infections",
                "reinfections",
            ]:
                aggregated_data[time_key]["locations"][location][key] += location_data[
                    key
                ]

    return aggregated_data
```

```python
new_covid_quarters = tranfrom_covid_days_to_quarters(covid_days)
```

```python
def transfrom_cancer_waiting_times(original_data):
    # Filter the data for the relevant periods (Q2 2020 to Q4 2021)
    cancer_waiting_times = [
        i
        for i in original_data
        if (i["Year"] == 2020 and i["Quarter"] > 1)
        or (i["Year"] > 2020 and i["Year"] < 2022)
    ]

    # Add 'Time' and 'Location' information to each entry in cancer_waiting_times
    for entry in cancer_waiting_times:
        # Concatenate year and quarter to create a unified 'Time' field
        entry["Time"] = str(entry["Year"]) + "-" + str(entry["Quarter"])

        # Extract 'Board_id' and lookup to assign the corresponding location name
        board_id = entry.get("Board_id")
        entry["Location"] = boards_code_lookup.get(board_id, "Unknown Location")

    # Initialize a dictionary to store aggregated cancer data grouped by 'Time'
    aggregated_cancer_data_per_time = {}

    # Iterate through the data to aggregate 'diagnosed' and 'treated' values per 'Time'
    for entry in cancer_waiting_times:
        # Retrieve or create the time key
        time_key = entry["Time"]

        # If this time key has not been added to the aggregated dictionary, add it and initialize count values
        if time_key not in aggregated_cancer_data_per_time:
            aggregated_cancer_data_per_time[time_key] = {
                "Total_diagnosed": 0,
                "Total_treated": 0,
            }

        # Iterate through each cancer type data, accumulating 'diagnosed' and 'treated'
        for cancer_type, cancer_data in entry["Cancer_types"].items():
            aggregated_cancer_data_per_time[time_key]["Total_diagnosed"] += cancer_data[
                "diagnosed"
            ]
            aggregated_cancer_data_per_time[time_key]["Total_treated"] += cancer_data[
                "treated"
            ]

    # Calculating and adding the treatment delay rate for each 'Time' in aggregated_cancer_data_per_time
    for time_key, data in aggregated_cancer_data_per_time.items():
        total_diagnosed = data["Total_diagnosed"]
        total_treated = data["Total_treated"]

        # Calculate the treatment delay rate and add it to the dictionary
        data["Treatment_delay_rate"] = (
            (total_diagnosed - total_treated) / total_diagnosed * 100
        )

    return aggregated_cancer_data_per_time
```

```python
new_cancer_waiting_times = transfrom_cancer_waiting_times(cancer_waiting_times)
```

```python
import matplotlib.pyplot as plt
import numpy as np


def visualization():
    fig, axs = plt.subplots(1, 2, figsize=(20, 8))

    for idx, exclude_first in enumerate([False, True]):
        times = sorted(list(new_cancer_waiting_times.keys()))
        if exclude_first:
            times = times[1:]

        # Extract the relevant data for the plots
        treatment_delay_rates = [
            new_cancer_waiting_times[time]["Treatment_delay_rate"] for time in times
        ]
        new_positives = [
            new_covid_quarters[time]["whole_scotland"]["new_positive"] for time in times
        ]
        first_infections = [
            new_covid_quarters[time]["whole_scotland"]["first_infections"]
            for time in times
        ]

        # Calculate correlation coefficients between treatment delay rates and Covid-19 metrics
        correlation_treatment_new_positive = np.corrcoef(
            treatment_delay_rates, new_positives
        )[0, 1]
        correlation_treatment_first_infections = np.corrcoef(
            treatment_delay_rates, first_infections
        )[0, 1]

        ax1 = axs[idx]

        # Label axes and plot the treatment delay rate
        ax1.set_xlabel("Time", fontsize=14)
        ax1.set_ylabel("Treatment Delay Rate (%)", color="tab:blue", fontsize=14)
        ax1.plot(
            times,
            treatment_delay_rates,
            color="tab:blue",
            marker="o",
            label="Treatment Delay Rate",
        )
        ax1.tick_params(axis="y", labelcolor="tab:blue", labelsize=12)
        ax1.tick_params(axis="x", rotation=45, labelsize=12)

        # Setup the second subplot for displaying Covid-19 metrics
        ax2 = ax1.twinx()
        ax2.set_ylabel("COVID-19 Metrics", color="tab:gray", fontsize=14)
        ax2.plot(
            times,
            new_positives,
            color="tab:red",
            marker="o",
            linestyle="--",
            label="New Positives",
        )
        ax2.plot(
            times,
            first_infections,
            color="tab:green",
            marker="o",
            linestyle="--",
            label="First Infections",
        )
        ax2.tick_params(axis="y", labelcolor="tab:gray", labelsize=12)

        # Add text box with correlation coefficients in the first subplot
        ax1.text(
            0.01,
            0.99,
            f"Corr(Treatment Delay, New Positives) = {correlation_treatment_new_positive:.2f}\nCorr(Treatment Delay, First Infections) = {correlation_treatment_first_infections:.2f}",
            transform=ax1.transAxes,
            verticalalignment="top",
            bbox=dict(boxstyle="round", alpha=0.1),
            fontsize=12,
        )

        ax1.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9), fontsize=12)
        ax2.legend(loc="upper left", bbox_to_anchor=(0.1, 0.8), fontsize=12)
        ax1.set_title(
            "Treatment Delay Rate and COVID-19 Metrics Over Time (Including 2020-1)"
            if not exclude_first
            else "Treatment Delay Rate and COVID-19 Metrics Over Time (Excluding 2020-1)",
            fontsize=16,
        )

    plt.tight_layout()
    plt.show()
```

```python
visualization()
```
