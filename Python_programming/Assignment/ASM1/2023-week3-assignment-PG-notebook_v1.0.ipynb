```python
# starting code that loads data, run this cell, but no need to understand it in depth
# this function just loads the data from files
def load_json_file_named(file_name):
    try: 
        loaded_data = []
        file_location = f"data/{file_name}"
        with open(file_location, 'r') as file: # or f"data/{file_name}" depending on your files
            loaded_data =  json.load(file)
    except OSError as e:
        print(f"Error. Does the file exist in this folder? {file_location}\n\n {e}")
    return loaded_data
```

```python
import matplotlib.pyplot as plt

# Extract date and first_infections data
dates = [str(entry["date"]) for entry in covid_days]
first_infections = [entry["whole_scotland"]["first_infections"] for entry in covid_days]

pp.pprint(dates[:5])
pp.pprint(first_infections)


# Plotting
plt.figure(figsize=(10, 4))
plt.plot(dates, first_infections, marker="o", linestyle="-", color="b")
plt.xticks(dates[::20], rotation=45)
plt.xlabel("Date")
plt.ylabel("First Infections")
plt.title("First Infections Over Time in Scotland")
plt.tight_layout()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
#  first graph:
## y: first-infection [whole time period(20200228-20220228)]
## x:location
total_infections = {location: 0 for location in covid_days[0]["locations"]}

for entry in covid_days:
    # Check if the date is within the desired range
    if 20200228 <= entry["date"] <= 20220228:
        for location, values in entry["locations"].items():
            total_infections[location] += values.get("first_infections", 0)

print(total_infections)

import matplotlib.pyplot as plt

# Extracting keys and values from the dictionary
locations = list(total_infections.keys())
values = list(total_infections.values())

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(locations, values, color="skyblue")
plt.xlabel("NHS Locations")
plt.ylabel("Number of Infections")
plt.title("Number of Infections by NHS Location")
plt.xticks(rotation=45, ha="right")  # Rotating x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
#  Second graph:
## y: total-death [whole time period(20200228-20220228)]
## x:location
total_death = {location: 0 for location in covid_days[0]["locations"]}

for entry in covid_days:
    # Check if the date is within the desired range
    if 20200228 <= entry["date"] <= 20220228:
        for location, values in entry["locations"].items():
            total_death[location] += values.get("new_deaths", 0)

print(total_death)

import matplotlib.pyplot as plt

# Extracting keys and values from the dictionary
locations = list(total_death.keys())
values = list(total_death.values())

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(locations, values, color="skyblue")
plt.xlabel("NHS Locations")
plt.ylabel("Number of Deaths")
plt.title("Number of Deaths by NHS Location")
plt.xticks(rotation=45, ha="right")  # Rotating x-axis labels for better visibility
plt.tight_layout()
plt.show()
```
